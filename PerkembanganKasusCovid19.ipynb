```python
# Import library datetime, pandas dan matplotlib
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Baca file Statistik_Perkembangan_Kasus_COVID19_Indonesia_Februari_2021.csv ke variable df, gunakan kolom 'Hari_ke' sebagai index
data = pd.read_csv('Statistik_Perkembangan_Kasus_COVID19_Indonesia_Februari_2021.csv', index_col='Hari_ke')
```

```python
# Hapus kolom yang terdapat pada list unused dari DataFrame df
unused = ['Kasus_Diperiksa_Baru_Harian', 'Jumlah_Kasus_Diperiksa', 'Jumlah_Spesimen_Diperiksa', 'Spesimen_Diperiksa_Baru_Harian', 'Jumlah_Negatif', 'PDP', 'ODP', 'Penambahan_Kasus_Terkonfirmasi_', 'Kasus_Sedang_Investigasi_Lapang', 'Jumlah_Spesimen_Diperiksa_Terki', 'Jumlah_Kasus_Diperiksa_Terkini', 'Spesimen_Diperiksa_Baru_Harian_', 'Jumlah_Kasus_Diperiksa_Harian_T', 'Hasil_Positif_Terkini', 'Hasil_Negatif_Terkini', 'PDP_Terkini', 'ODP_Terkini', 'Suspek_Terkini', 'Jumlah_Provinsi_Terdampak',  'Jumlah_KabKot_Terdampak', 'Pembaruan_Terakhir', 'FID']
data.drop(unused, axis=1, inplace=True)
```

```python
# Ubah tipe data kolom Tanggal ke datetime
data['Tanggal'] = pd.to_datetime(data['Tanggal'])
```

```python
# Urutkan data berdasarkan indexnya secara menaik
data.sort_index(inplace=True)
```

*#Kasus baru = #sembuh + #meninggal + #dirawat

```python
# Your code here...
data.at[263,'Jumlah_Kasus_Baru_per_Hari'] = 4798
data.at[263,'Jumlah_Kasus_Sembuh_per_Hari'] = 4265
data.at[263,'Jumlah_Kasus_Meninggal_per_Hari'] = 97
data.at[263,'Jumlah_Kasus_Dirawat_per_Hari'] = 436
data.loc[263]
```

```python
data.at[262,'Jumlah_Kasus_Kumulatif'] = data.at[263,'Jumlah_Kasus_Kumulatif'] - data.at[263,'Jumlah_Kasus_Baru_per_Hari']
data.at[262,'Jumlah_Kasus_Baru_per_Hari'] = data.at[262,'Jumlah_Kasus_Kumulatif'] - data.at[261,'Jumlah_Kasus_Kumulatif']

data.at[262,'Jumlah_Pasien_Sembuh'] = data.at[263,'Jumlah_Pasien_Sembuh'] - data.at[263,'Jumlah_Kasus_Sembuh_per_Hari']
data.at[262,'Jumlah_Kasus_Sembuh_per_Hari'] = data.at[262,'Jumlah_Pasien_Sembuh'] - data.at[261,'Jumlah_Pasien_Sembuh']

data.at[262,'Jumlah_Pasien_Meninggal'] = data.at[263,'Jumlah_Pasien_Meninggal'] - data.at[263,'Jumlah_Kasus_Meninggal_per_Hari']
data.at[262,'Jumlah_Kasus_Meninggal_per_Hari'] = data.at[262,'Jumlah_Pasien_Meninggal'] - data.at[261,'Jumlah_Pasien_Meninggal']

data.at[262,'Jumlah_Pasien_Dalam_Perawatan'] = data.at[262,'Jumlah_Kasus_Kumulatif'] - data.at[262,'Jumlah_Pasien_Sembuh'] - data.at[262,'Jumlah_Pasien_Meninggal']
data.at[262,'Jumlah_Kasus_Dirawat_per_Hari'] = abs(data.at[262,'Jumlah_Pasien_Dalam_Perawatan'] - data.at[261,'Jumlah_Pasien_Dalam_Perawatan'])

data.at[262,'Persentase_Pasien_Sembuh'] = (data.at[262,'Jumlah_Pasien_Sembuh'] / data.at[262,'Jumlah_Kasus_Kumulatif']) * 100
data.at[262,'Persentase_Pasien_Meninggal'] = (data.at[262,'Jumlah_Pasien_Meninggal'] / data.at[262,'Jumlah_Kasus_Kumulatif']) * 100
data.at[262,'Persentase_Pasien_dalam_Perawat'] = (data.at[262,'Jumlah_Pasien_Dalam_Perawatan'] / data.at[262,'Jumlah_Kasus_Kumulatif']) * 100
data.loc[262]
```

```python
# Tampilkan kembali data hari ke-262 dan data hari ke-263
data.loc[261:263]
```

```python
# Berapa jumlah pertambahan kasus pada bulan Januari 2021
pertambahan = data['Jumlah_Kasus_Baru_per_Hari'].loc[(data['Tanggal'] >= 'January 2021') & (data['Tanggal'] < 'February 2021')].sum()
print(f'Pertambahan kasus pada bulan Januari 2021 adalah {int(pertambahan)} kasus')
```

```python
# Buatlah sebuah grafik yang menampilkan persentase kasus kumulatif (dalam perawatan, sembuh, meninggal) pasien Covid19 pada tanggal 18 November 2020
label = 'perawatan', 'sembuh', 'meninggal'
kumulatif = data.loc[262]
total = [kumulatif['Persentase_Pasien_dalam_Perawat'],kumulatif['Persentase_Pasien_Sembuh'],kumulatif['Persentase_Pasien_Meninggal']]
plt.pie(total,labels=label,autopct='%0.1f%%')
plt.title('Grafik Kumulatif Pada Tanggal 18 November 2020')
plt.show()
```

```python
# Buatlah sebuah grafik yang menampilkan jumlah pertambahan kasus per bulan dari bulan Maret 2020 sampai Februari 2021
months = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
marFeb = {}

year = 2020
for i in range(2, 12):
  marFeb[f'{months[i]}, {year}'] = int(data[(data['Tanggal'].dt.year == year) & (data['Tanggal'].dt.month == i+1 )]['Jumlah_Kasus_Baru_per_Hari'].sum())

year = 2021
for i in range(2):
  marFeb[f'{months[i]}, {year}'] = int(data[(data['Tanggal'].dt.year == year) & (data['Tanggal'].dt.month == i+1 )]['Jumlah_Kasus_Baru_per_Hari'].sum())
marFeb
fig = plt.figure(figsize=(10,5))
plt.barh(list(marFeb.keys()), list(marFeb.values()), color='#006400')
plt.xlabel("Jumlah Kasus")
plt.ylabel("Bulan")
plt.title("Pertambahan Kasus per Bulan Dari Bulan Maret 2020 sampai Februari 2021")
plt.show()
```

```python
# Buatlah sebuah grafik kasus kumulatif Covid19 selama tahun 2020
months = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']
marFeb = {}

year = 2020
listSum = []
sum_ = 0
for i in range(2, 12):
  marFeb[f'{months[i]}, {year}'] = int(data[(data['Tanggal'].dt.year == year) & (data['Tanggal'].dt.month == i+1 )]['Jumlah_Kasus_Baru_per_Hari'].sum())
  sum_ += marFeb[f'{months[i]}, {year}']
  listSum.append(sum_)
fig = plt.figure(figsize=(10,5))
plt.barh(list(marFeb.keys()), listSum, color='#006400')
plt.xlabel("Jumlah Kasus")
plt.ylabel("Bulan")
plt.title("Kasus Kumulatif Covid19 Selama Tahun 2020")
plt.show()
```
